```python

%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime
```

```python
densenet121 = pd.read_csv('history_DenseNet121_July 30, 2018.csv',';') # ALTERAR
densenet169 = pd.read_csv('history_DenseNet169_July 30, 2018.csv',';')
densenet201 = pd.read_csv('history_DenseNet201_July 31, 2018.csv',';')
inceptionv3 = pd.read_csv('history_InceptionV3_July 30, 2018.csv',';')
resnet50 = pd.read_csv('history_ResNet50_July 31, 2018.csv',';')
vgg16 = pd.read_csv('history_VGG16_July 30, 2018.csv',';')
vgg19 = pd.read_csv('history_VGG19_July 30, 2018.csv',';')
xception = pd.read_csv('history_Xception_July 30, 2018.csv',';')
```

## Accuracy plot

```python
accuracies = np.vstack([densenet121.iloc[:,1],
                        densenet169.iloc[:,1],
                        densenet201.iloc[:,1],
                        inceptionv3.iloc[:,1],
                        resnet50.iloc[:,1],
                        vgg16.iloc[:,1],
                        vgg19.iloc[:,1],
                        xception.iloc[:,1]]).transpose()
```

```python
#accuracies
```

```python
accuracy = pd.DataFrame(accuracies, columns=['densenet121',
                                             'denset169',
                                             'densenet201',
                                             'inceptionv3',
                                             'resnet50',
                                             'vgg16',
                                             'vgg19',
                                             'xception'])
accuracy
```

```python
fig = plt.figure(figsize=(12,8), dpi=100, facecolor='w', edgecolor='k')
plt.plot(accuracy, linewidth=2, markersize=12)
plt.title('Train Accuracy')
plt.legend(['densenet121',
            'denset169',
            'densenet201',
            'inceptionv3',
            'resnet50',
            'vgg16',
            'vgg19',
            'xception'])
fig.savefig('Train_Accuracy_all_models_{}.jpg'.format(datetime.date.today().strftime("%B %d, %Y")))
#plt.close(fig)
```

## Loss plot

```python
losses = np.vstack([densenet121.iloc[:,2],
                    densenet169.iloc[:,2],
                    densenet201.iloc[:,2],
                    inceptionv3.iloc[:,2],
                    resnet50.iloc[:,2],
                    vgg16.iloc[:,2],
                    vgg19.iloc[:,2],
                    xception.iloc[:,2]]).transpose()
```

```python
loss = pd.DataFrame(losses, columns=['densenet121',
                                             'denset169',
                                             'densenet201',
                                             'inceptionv3',
                                             'resnet50',
                                             'vgg16',
                                             'vgg19',
                                             'xception'])
```

```python
fig = plt.figure(figsize=(12,8), dpi=100, facecolor='w', edgecolor='k')
plt.plot(loss, linewidth=2, markersize=12)
plt.title('Train Loss')
plt.legend(['densenet121',
            'denset169',
            'densenet201',
            'inceptionv3',
            'resnet50',
            'vgg16',
            'vgg19',
            'xception'])
fig.savefig('Train_Loss_all_models_{}.jpg'.format(datetime.date.today().strftime("%B %d, %Y")))
#plt.close(fig)
```
